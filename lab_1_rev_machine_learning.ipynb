Identify columns that over 85% missing values

```python
categorical['MAILCODE'] = categorical['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
```

```python
drop_columns = []
def na_to_remove(df, threshold):    
    
    df_2 = df.copy()
    cols = df_2.columns
    rows = df_2.shape[0]
    
    for col in cols:
        na = df_2[col].isna().sum()
        if na > (rows * threshold/100):
            drop_columns.append(col)
        else: 
            pass
    
    return drop_columns
```

```python
na_values_over_85 = na_to_remove(categorical, 85)
```

```python
na_values_over_85
```

Remove those columns from the dataframe

```python
for i in na_values_over_85:
    categorical = categorical.drop(i, axis = 1)
```
